```
max=int(input("enter maximum value"))
print("palindrome numbers between 1 to %d are",max)
for num in range(1,max+1):
  temp=num
  rev=0
  while(temp>0):
    rem=temp%10
    rev=(rev*10)+rem
    temp=temp//10
    if num==rev:
       print(num,end=' ')
```

```
print("Armstrong numbers are:")
for num in range(1,1001):
  temp=num
  sum=0
  while temp>0:
    digit=temp%10
    sum=sum+digit**3
    temp=temp//10
    if num==sum:
      print(sum,end=" ")
```
